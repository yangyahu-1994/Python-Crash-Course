
### 2.2.1 变量的命名和使用
#### 变量名只能包含字母、数字和下划线。变量名可以字母或下划线打头，但不能以数字打头。message_1/1_message
#### 变量名不能包含空格，但可使用下滑线来分割其中的单词。greeting_message/greeting message
#### 不要将Python关键字和函数名用作变量名。
#### 变量名应既简短又具有描述性。
#### 慎用小写字母l和大写字母O，因为他们可能被人错看成数字1和0.

### 2.2.2 使用变量时避免命名错误

```python
message = "Hello Python Crash Course reader!"
print(message)
## Python解释器不会对代码做拼写检查，但要求变量名的拼写一致。
mesage = "Hello Python Crash Course reader!"
print(mesage)
```

## 2.3 字符串
#### 在python中，用引号括起来的都是字符串，其中的引号可以是单引号，也可以是双引号。
### 2.3.1 使用方法修改字符串的大小写

```python
# name.py
name = "ada lovelace"
print(name.title())
name = "Ada Lovelace"
print(name.upper())
print(name.lower())
```

### 2.3.2 合并（拼接）字符串

```python
fist_name = "ada"
last_name = "lovelace"
full_name = fist_name + " " + last_name
print(full_name)
## python使用加号(+)来合并字符串。
fist_name = "ada"
last_name = "lovelace"
full_name = fist_name + " " + last_name
print("Hello, " + full_name.title() + "!")
fist_name = "ada"
last_name = "lovelace"
full_name = fist_name + " " + last_name
message = "Hello, " + full_name.title() + "!"
print(message)
```

### 2.3.3 使用制表符和换行符来添加空白

```python
print("python")
print("\tpython")
print("Languages:\npython\nC\nJavaScript")
# 要在字符串中添加制表符，可使用字符组合\t
# 要在字符串中添加换行符，可使用字符组合\n
print("Languages:\n\tpython\n\tC\n\tJavaScript")
```
